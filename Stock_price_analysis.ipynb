## Project Objective

The objective of this project is to analyze historical financial data of Apple Inc. to identify time-based trends and patterns in stock prices.  
Through time-series visualization and trend analysis, the project aims to support better business and investment decision-making.


```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\jasmin's skillinfytech project\stock_price_analysis\AAPL.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

The date column was converted to datetime format to enable accurate time-series analysis.

```python
df.isnull().sum()
```

## Time-Series Visualization

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'],df['Close'])
plt.title("Apple Stock Closing Price Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.show()
```

The stock shows a strong long-term upward trend, indicating consistent growth over the years despite short-term market fluctuations.


## Month-on-Month Trend Analysis

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')['Close'].mean()
```

```python
plt.figure(figsize=(10,5))
monthly_avg.plot()
plt.title('Month-on-Month Average Closing Price')
plt.xlabel('Month')
plt.ylabel('Average Closing Price (USD)')
plt.show()
```

The monthly trend smooths daily volatility and reveals sustained growth phases along with temporary corrections during market downturns.


## Moving Average Trend 

```python
## 50 - day & 200-day moving averages 
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_200'] = df['Close'].rolling(window=200).mean()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'],df['Close'],label='Close Price')
plt.plot(df['Date'],df['MA_50'],label='50-day MA')
plt.plot(df['Date'],df['MA_200'],label='200-day MA')
plt.legend()
plt.title('Apple Stock Price with Moving Averages')
plt.show()
```

The moving averages confirm a long-term bullish trend, where the closing price consistently follows the 50-day and 200-day averages, signaling overall market strength.
